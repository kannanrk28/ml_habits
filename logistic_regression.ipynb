```python
import math

def sigmoid(z):
    """Compute sigmoid probability for a real-valued input."""
    return 1 / (1 + math.exp(-z))

def predict_class(z):
    """Predict binary class (0 or 1) based on sigmoid probability."""
    prob = sigmoid(z)
    return 1 if prob >= 0.5 else 0

def evaluate(y_true, y_pred):
    """Evaluate predictions and return accuracy, FP, and FN."""
    correct = sum(yt == yp for yt, yp in zip(y_true, y_pred))
    total = len(y_true)

    false_positives = sum((yt == 0 and yp == 1) for yt, yp in zip(y_true, y_pred))
    false_negatives = sum((yt == 1 and yp == 0) for yt, yp in zip(y_true, y_pred))

    accuracy = correct / total

    return {
        "accuracy": accuracy,
        "false_positives": false_positives,
        "false_negatives": false_negatives
    }

def display_report(y_true, y_pred, z_values):
    """Display probabilities, predictions, and evaluation metrics."""
    probabilities = [sigmoid(z) for z in z_values]
    predictions = [predict_class(z) for z in z_values]
    metrics = evaluate(y_true, predictions)

    print("Probabilities    :", probabilities)
    print("Predicted Classes:", predictions)
    print(f"Accuracy         : {metrics['accuracy'] * 100:.2f}%")
    print("False Positives  :", metrics['false_positives'])
    print("False Negatives  :", metrics['false_negatives'])


# Test data
z_values = [2.5, -1.2, 0.0, 3.1, -2.8]
y_true   = [1, 0, 1, 1, 0]

# Run
display_report(y_true, None, z_value)
```
